```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/kaggle/input/insurance/insurance_data.csv')
X = df.iloc[:,0].values
y = df.iloc[:,-1].values
```

```python
plt.scatter(X,y,color='red')
plt.show()
```

```python
df.shape
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X.reshape(-1,1),y.reshape(-1,1),train_size=0.9)
```

```python
X_test
```

## Logistic Regression 

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(X_train,y_train)
```

## Prediction of Model

```python
lr.predict(X_test)
```

```python
lr.score(X_test,y_test)
```

```python
lr.predict_proba(X_test)
```

```python
#First column show probability of a person is going to NOT bought insurance
#Second column show probability of a person is going to bought insurance
```

```python
X_new = np.linspace(10, 100, 27).reshape(-1, 1)
y_proba = lr.predict_proba(X_new)
plt.scatter(X_train,y_train,color='red')
plt.plot(X_new, y_proba[:, 1], "g-" )
plt.plot(X_new, y_proba[:, 0], "b--")
```
