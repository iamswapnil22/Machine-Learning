## Implementation of Ridge Machine Learning Model

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('/kaggle/input/melbourne-housing-market/Melbourne_housing_FULL.csv')
dataset.head()
```

```python
dataset.nunique()
```

## Linear Regression

```python
cols_to_use = ['Suburb','Rooms','Type','Method','SellerG','Regionname',
              'Propertycount','Distance','CouncilArea','Bedroom2','Bathroom','Car','Landsize','BuildingArea','Price']
dataset = dataset[cols_to_use]
dataset.head()
```

```python
dataset.shape
```

```python
dataset.isna().sum()
```

```python
cols_to_zero = ['Propertycount','Distance','CouncilArea','Bedroom2','Bathroom','Car']
dataset[cols_to_zero] = dataset[cols_to_zero].fillna(0)
dataset.isna().sum()
```

```python
dataset['Landsize'] = dataset['Landsize'].fillna(dataset.Landsize.mean())
dataset['BuildingArea'] = dataset['BuildingArea'].fillna(dataset.BuildingArea.mean())
dataset['Landsize'] = dataset['Landsize'].fillna(dataset.Landsize.mean())
dataset.isna().sum()
```

```python
dataset.dropna(inplace=True)
dataset.isna().sum()
```

```python
dataset = pd.get_dummies(dataset,drop_first=True)
dataset.head()
```

```python
X = dataset.drop('Price',axis=1)
y = dataset['Price']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.3 ,random_state = 2)
```

### Linear Regression

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
reg = lr.fit(X_train,y_train)
```

```python
reg.score(X_test,y_test)
```

```python
reg.score(X_train,y_train)
```

### Lasso Regression

```python
from sklearn.linear_model import Lasso
ls = Lasso(alpha=50,max_iter=100,tol=0.1)
reg2 = ls.fit(X_train,y_train)
```

```python
reg2.score(X_test,y_test)
```

```python
reg2.score(X_train,y_train)
```

### Ridge Regression

```python
from sklearn.linear_model import Ridge
rd = Ridge(alpha=50,max_iter=100,tol=0.1)
reg3 = rd.fit(X_train,y_train)
```

```python
reg3.score(X_test,y_test)
```

```python
reg3.score(X_train,y_train)
```
